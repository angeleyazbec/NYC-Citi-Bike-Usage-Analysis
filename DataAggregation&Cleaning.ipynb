```python
# importing dependencies
import pandas as pd
import glob
import numpy as np
from math import sin, cos, sqrt, atan2, radians
import datetime
```

```python
#using glob to obtain the list of files
folder_path = 'Data'
file_list = glob.glob(folder_path + "/*.csv")
file_list
```

```python
# reading in multiple files using Map
li_mapper = map(lambda filename: pd.read_csv(filename, index_col= None), file_list)
li_2 = list(li_mapper)
df = pd.concat(li_2, axis=0, ignore_index=True)
df
```

```python
#filling any blank cells with NaN
clean_df = df.replace(r'^\s*$', np.nan, regex=True)

clean_df
```

```python
#dropping any rows with incomplete data
clean_df2=clean_df.dropna(axis=0, how="any")

clean_df2
```

```python
R = 6373.0
for index, row in clean_df2.iterrows():
    lat1 = radians(row["start_lat"])
    lon1 = radians(row["start_lng"])
    lat2 = radians(row["end_lat"])
    lon2 = radians(row["end_lng"])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance_km = R * c
    distance_m_unrounded = distance_km * 0.62137
    distance_m = round(distance_m_unrounded, 3)
    clean_df2.loc[index, "distance_m"] = distance_m
    clean_df2.loc[index, "distance_m_unrounded"] = distance_m_unrounded
```

```python
#writing the file to a csv to future use
clean_df2.to_csv("Data/citi_bikes.csv", index=False, header=True)
```
